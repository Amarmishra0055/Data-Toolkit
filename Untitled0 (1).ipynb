1. **What is NumPy, and why is it widely used in Python?**  
 - NumPy is a powerful Python library used for numerical and scientific computing. It provides fast operations on arrays and matrices, and supports a wide range of mathematical functions. It is widely used because it performs faster computations than regular Python lists and supports operations like linear algebra, Fourier transforms, and random number generation, which are essential for data analysis and machine learning tasks.

2. **How does broadcasting work in NumPy?**  
 - Broadcasting is a feature in NumPy that allows operations between arrays of different shapes without copying data. When performing operations on two arrays, NumPy automatically expands the smaller array to match the shape of the larger one. This avoids loops and improves performance. For example, adding a scalar to an array uses broadcasting, as the scalar is applied to every element efficiently.

3. **What is a Pandas DataFrame?**  
 - A Pandas DataFrame is a two-dimensional, tabular data structure in Python that looks like a spreadsheet or SQL table. It consists of rows and columns, where each column can hold different data types. DataFrames are widely used in data analysis for organizing, filtering, and transforming data. They offer powerful tools for handling missing values, grouping, merging, and reshaping data.

4. **Explain the use of the groupby() method in Pandas.**  
 - The groupby() method in Pandas is used to split data into groups based on one or more columns. It allows you to apply functions like sum, mean, or count to each group separately. This is especially useful for analyzing categories in large datasets, such as finding the average sales by product or total revenue by region, helping in summarizing and understanding the data better.

5. **Why is Seaborn preferred for statistical visualizations?**  
 - Seaborn is preferred for statistical visualizations because it is built on top of Matplotlib and offers simpler syntax and better default themes. It supports visualizing relationships between variables using plots like heatmaps, box plots, and pair plots. It also integrates well with Pandas DataFrames and makes it easier to plot distributions, trends, and category comparisons using fewer lines of code.

6. **What are the differences between NumPy arrays and Python lists?**  
 - NumPy arrays are more efficient for numerical operations compared to Python lists. They use less memory, support element-wise operations, and are much faster due to their fixed type and use of underlying C code. Lists are more flexible with mixed data types but are slower and lack mathematical capabilities. For data analysis or matrix calculations, NumPy arrays are a better choice.

7. **What is a heatmap, and when should it be used?**  
 - A heatmap is a graphical representation of data where values are shown using colors. It’s useful for visualizing the strength of relationships between variables, especially in correlation matrices. Heatmaps are commonly used in data exploration to detect patterns, outliers, or trends in large datasets, such as identifying high-performing areas or finding clusters in machine learning models.

8. **What does the term “vectorized operation” mean in NumPy?**  
 - A vectorized operation in NumPy refers to performing operations on entire arrays without using loops. Instead of processing elements one by one, NumPy applies operations to all elements at once using optimized C-based functions. This approach improves speed, readability, and reduces errors, making it ideal for numerical and scientific computing tasks that involve large datasets.

9. **How does Matplotlib differ from Plotly?**  
 - Matplotlib is a basic 2D plotting library suitable for static visualizations, while Plotly is an interactive plotting library that allows zooming, panning, and hovering. Plotly is ideal for dashboards and presentations because of its rich, web-friendly visuals. Matplotlib offers more control for custom plots but requires more code. Plotly simplifies creating attractive, interactive charts using fewer commands.

10. **What is the significance of hierarchical indexing in Pandas?**  
 - Hierarchical indexing, or multi-level indexing, allows Pandas to handle data with more than one index level. This is useful for working with higher-dimensional data in a 2D DataFrame. It lets you group and analyze data by multiple keys, such as time and location. It also supports reshaping and slicing complex datasets, offering powerful tools for advanced data analysis.

11. **What is the role of Seaborn’s pairplot() function?**  
 - Seaborn’s pairplot() is used to visualize pairwise relationships between numeric variables in a dataset. It automatically creates scatter plots for each variable combination and histograms or KDE plots on the diagonal. This is useful for quickly understanding the distribution and correlation of features, making it a common first step in exploratory data analysis.

12. **What is the purpose of the describe() function in Pandas?**  
 - The describe() function in Pandas provides a quick summary of numeric columns in a DataFrame. It returns statistics like count, mean, standard deviation, minimum, maximum, and quartiles. This helps analysts understand the distribution and spread of the data, identify outliers, and detect errors early in the analysis process. It is a key step in data exploration.

13. **Why is handling missing data important in Pandas?**  
 - Handling missing data is important because unhandled gaps can lead to incorrect analysis, errors in computations, or biased results. Pandas provides tools to detect, fill, or remove missing values using functions like isnull(), fillna(), and dropna(). Properly dealing with missing data ensures accuracy and consistency, especially when building machine learning models or generating reports.

15. **How does NumPy handle multidimensional arrays?**

It allows indexing, slicing, and reshaping across dimensions easily. Operations like matrix multiplication or axis-specific functions work efficiently on multidimensional data. This makes NumPy especially useful for tasks in image processing, physics, and deep learning.

16. **What is the role of Bokeh in data visualization?**

17. **Explain the difference between apply() and map() in Pandas.**  
 - In Pandas, map() is used mainly on Series (one column) to apply a function element-wise. It’s simple and works best for transformations or replacements. apply() is more powerful and can be used on both Series and DataFrames. It can apply functions to rows or columns, making it ideal for complex operations. Use map() for straightforward changes and apply() for custom logic across rows or columns.

18. **What are some advanced features of NumPy?**  
 - NumPy offers advanced features like broadcasting, masking, vectorized operations, memory mapping of large arrays, integration with C/Fortran code, and linear algebra tools. It also includes Fourier transforms, random number generation, and complex indexing techniques. These features make NumPy powerful for scientific computing and numerical analysis, allowing developers to build efficient, high-performance data processing pipelines.

19. **How does Pandas simplify time series analysis?**

You can easily convert strings to datetime, set time columns as indexes, and perform operations like aggregating monthly data or calculating moving averages. This makes time series analysis, such as forecasting or tracking changes over time, more intuitive and powerful using simple code.

20. **What is the role of a pivot table in Pandas?**  
 - A pivot table in Pandas is used to summarize, group, and rearrange data. It allows you to compute statistics like count, mean, or sum based on one or more keys. It's particularly helpful when analyzing large datasets or producing reports. By pivoting, you can view trends, spot patterns, or compare categories with minimal code, similar to Excel's pivot functionality.

21. **Why is NumPy’s array slicing faster than Python’s list slicing?**  
 - NumPy arrays are stored in contiguous memory blocks and use fixed data types, allowing fast element access and memory efficiency. Python lists, on the other hand, store references to objects and support mixed types, which adds overhead. Because of this, NumPy slicing avoids the overhead of object resolution and executes faster, especially for large-scale data operations in numerical computing.

```python
# 1. How do you create a 2D NumPy array and calculate the sum of each row
import numpy as np

arr = np.array([[1, 2, 3], [4, 5, 6]])
row_sums = arr.sum(axis=1)
print("2D Array:\n", arr)
print("Sum of each row:", row_sums)
```

```python
# 2. Write a Pandas script to find the mean of a specific column in a DataFrame
import pandas as pd

df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Score': [85, 90, 78]
})
mean_score = df['Score'].mean()
print("Mean of 'Score' column:", mean_score)
```

```python
# 3. Create a scatter plot using Matplotlib
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [5, 7, 4, 6, 8]
plt.scatter(x, y, color='red')
plt.title('Scatter Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
# 4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap
import seaborn as sns

data = pd.DataFrame({
    'A': [1, 2, 3, 4],
    'B': [4, 5, 6, 7],
    'C': [7, 6, 5, 4]
})
correlation = data.corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# 5. Generate a bar plot using Plotly
import plotly.express as px

data = pd.DataFrame({
    'Fruits': ['Apples', 'Bananas', 'Cherries'],
    'Quantity': [10, 15, 7]
})
fig = px.bar(data, x='Fruits', y='Quantity', title='Fruit Quantity Bar Plot')
fig.show()
```

```python
# 7. Write a program to perform element-wise multiplication of two NumPy arrays
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
result = arr1 * arr2
print("Element-wise multiplication:", result)
```

```python
# 8. Create a line plot with multiple lines using Matplotlib
x = [1, 2, 3, 4, 5]
y1 = [2, 3, 4, 5, 6]
y2 = [3, 4, 2, 6, 7]
plt.plot(x, y1, label='Line 1')
plt.plot(x, y2, label='Line 2')
plt.legend()
plt.title('Multiple Line Plot')
plt.show()
```
